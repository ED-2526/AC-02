```python
import re
from collections import Counter

errors_maxims = 35

# Leer el archivo con encoding latin-1
with open('30 pixels\\proves_30_pixels.txt', 'r', encoding='latin-1') as file:
    content = file.read()

# Extraer todas las líneas de "Males classificacions"
error_pattern = r'ID: (\d+)  --> Real: ([^,]+),  Pred: ([^\n]+)'
errors = re.findall(error_pattern, content)

# Contar cuántas veces falla cada ID
id_counter = Counter()
for error in errors:
    id_num = int(error[0])
    id_counter[id_num] += 1

# Filtrar IDs con exactamente 35 errores
errors_30_pixels = []
for id_num, count in id_counter.items():
    if count == errors_maxims:
        errors_30_pixels.append(id_num)

print(f"IDs con exactamente {errors_maxims} errores ({len(errors_30_pixels)} encontrados):")
print("-" * 50)
```

```python
import re
from collections import Counter

errors_maxims = 28

# Leer el archivo con encoding latin-1
with open('20 pixels\\proves_20_pixels.txt', 'r', encoding='latin-1') as file:
    content = file.read()

# Extraer todas las líneas de "Males classificacions"
error_pattern = r'ID: (\d+)  --> Real: ([^,]+),  Pred: ([^\n]+)'
errors = re.findall(error_pattern, content)

# Contar cuántas veces falla cada ID
id_counter = Counter()
for error in errors:
    id_num = int(error[0])
    id_counter[id_num] += 1



# Filtrar IDs con exactamente 28 errores
errors_20_pixels = []
for id_num, count in id_counter.items():
    if count == errors_maxims:
        errors_20_pixels.append(id_num)

errors_20_pixels.sort()

print(f"IDs con exactamente {errors_maxims} errores ({len(errors_20_pixels)} encontrados):")
print("-" * 50)

```

```python
comunes = [x for x in errors_30_pixels if x in errors_20_pixels]

for element in comunes:
    print(f"ID: {element}")

print("Total d' elements comuns:", len(comunes))
#print(f"Elementos comunes: {comunes}")
```
